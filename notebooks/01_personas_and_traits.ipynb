# 01 · Personas & traits

The project induces **personas** via system prompts and extracts steering vectors
for **8 traits**. This notebook explores both using the package's loaders.

```python
from persona_steering.personas import load_all_personas
from persona_steering.config import Trait

personas = load_all_personas()
print(f'{len(personas)} personas loaded from data/personas/')
for p in personas:
    print(f'  {p.slug:24s} {p.name}')
```

Each `PersonaConfig` carries a `name`, `slug`, and **5 system-prompt variants**
(used for robust extraction — the same persona phrased five ways).

```python
p = personas[0]
print('slug      :', p.slug)
print('name      :', p.name)
print(f'# variants : {len(p.system_prompt_variants)}')
print('variant[0]:')
print(p.default_system_prompt[:500])
```

## Traits

The 8 traits are an `Enum`. Each has an instruction-variant dataset: 5 positive /
negative instruction pairs + 100 shared questions.

```python
for t in Trait:
    print(f'  {t.name:14s} = {t.value!r}')
```

```python
from persona_steering.data import load_trait_dataset

ds = load_trait_dataset(Trait.ASSERTIVENESS)
pair = ds.instruction_variants[0]
print('POSITIVE instruction:\n ', pair.positive_instruction)
print('\nNEGATIVE instruction:\n ', pair.negative_instruction)
print(f'\n# shared questions: {len(ds.questions)}')
print('example question:', ds.questions[0])
```

**Why this design?** Asking the *same* question under opposing instructions controls
for content, so the contrastive vector `mean(pos) - mean(neg)` isolates the *trait*
signal rather than topic differences. ➡️ Next:
